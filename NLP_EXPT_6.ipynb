```python
import spacy
from nltk.stem import PorterStemmer

# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Apple is looking at buying a startup in Bengaluru for $1 billion."

# Process text
doc = nlp(text)

print("Original Text:")
print(text)
```

```python
# 1. TOKENIZATION
print("Tokenization:")
tokens = [token.text for token in doc]
print(tokens)
```

```python
# 2. STEMMING (spaCy doesn't support stemming directly, so use NLTK)
ps = PorterStemmer()
stemmed_words = [ps.stem(token.text) for token in doc if token.is_alpha]
print("\nStemming:")
print(stemmed_words)
```

```python
lemmas = [token.lemma_ for token in doc]
print("\nLemmatization:")
print(lemmas)
```

```python
print("\nPOS Tagging:")
for token in doc:
    print(token.text, "->", token.pos_)
```

```python
# 5. NAMED ENTITY RECOGNITION (NER)
print("\nNamed Entity Recognition:")
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

```python
# 6. STOP WORD REMOVAL
stopwords_removed = [token.text for token in doc if not token.is_stop and token.is_alpha]
print("\nStop Word Removal:")
print(stopwords_removed)
```

```python
import spacy

# Load ML-based spaCy model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "The children are playing football in the park."

# Process text using trained ML model
doc = nlp(text)

print("Word\tPOS Tag\tLemma")
print("---------------------------")

# POS Tagging and Lemmatization
for token in doc:
    print(token.text, "\t", token.pos_, "\t", token.lemma_)
```

```python
!pip install emoji
import re
import emoji

text = "Heeey!!! This is soooo cooool üòçüòç $$$"

print("Original Text:")
print(text)

# Lowercase
text = text.lower()

# Remove emojis
text = emoji.replace_emoji(text, replace='')

# Remove special characters
text = re.sub(r'[^a-z\s]', '', text)

# Reduce ALL repeated characters to ONE
text = re.sub(r'(.)\1+', r'\1', text)

# Restore valid 'oo' (optional but needed)
text = re.sub(r'\bcol\b', 'cool', text)

# Remove extra spaces
text = re.sub(r'\s+', ' ', text).strip()

print("\nCleaned Text:")
print(text)
```
